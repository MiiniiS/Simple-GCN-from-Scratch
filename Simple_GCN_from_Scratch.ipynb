```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
x=np.array([[2,3],
           [4,5],
           [8,9],
           [1,2]])
a=np.array([[1,1,0,0],
           [1,1,1,0],
           [0,1,1,1],
           [0,0,1,1]])
def normalization (a):
  d=np.diag(np.sum(a,axis=1))
  d_inv=np.linalg.inv(np.sqrt(d))
  a_hat=d_inv@a@d_inv
  return a_hat
def GCN_layer(a_hat,x,w):
  return np.maximum(0,a_hat@x@w)

input_dim=x.shape[1]
hidden_dim=4

w1=np.random.randn(input_dim,hidden_dim)
a_hat=normalization(a)
h1=GCN_layer(a_hat,x,w1)
print(h1)

w2=np.random.randn(hidden_dim,2)
h2=GCN_layer(a_hat,h1,w2)
print("H2 WILL BE", h2)
def plot_graph(A):
    G = nx.from_numpy_array(a)
    pos = nx.spring_layout(G)  # Layout for visualization
    plt.figure(figsize=(6, 4))
    nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=1000, edge_color='gray', font_size=12)
    plt.title("Graph Structure")
    plt.show()

plot_graph(a)
```
